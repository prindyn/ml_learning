```python
import numpy as np
import sympy as sp
import matplotlib.pyplot as plt
```

### Домашні завдання на відпрацювання поняття функції та її похідної

1. Задана функція залежності оцінки від кількості годин, витрачених на навчання: $ f(x) = 10x - 5 $.
  1. З допомогою Python обчисліть похідну цієї функції.
  2. Використовуйте бібліотеку `matplotlib` для побудови графіків функції та її похідної в діапазоні значень від 0 до 10 годин. Проаналізуйте, як зміна кількості годин вивчення впливає на оцінку.

```python
def f(x):
    return 10 * x - 5

def f_prime(x):
    return [10] * len(x)

x = np.linspace(0, 10, 100)

plt.figure(figsize=(10, 5))
plt.plot(x, f(x), label="f(x) = 10x - 5", color='b')
plt.plot(x, f_prime(x), label="f'(x) = 10", color='g', linestyle='--')

plt.legend()
plt.show()
```

**Висновок**

- У кожній точці функції оцінка збільшується з тою ж швидкістю.

2. Використовуючи Python і `SymPy`, обчисліть похідні для функцій $ f(x) = \sin(x) $ та $ f(x) = \ln(x) $. Зобразіть ці функції та їхні похідні за допомогою `matplotlib`.

```python
x = sp.symbols('x')

f1 = sp.sin(x)
f2 = sp.ln(x)

f1_prime = sp.diff(f1)
f2_prime = sp.diff(f2)

print(f1_prime, f2_prime)

x_vals = np.linspace(0, 10, 400)

f1_vals = np.sin(x_vals)
f2_vals = np.log(x_vals)

f1_prime_vals = np.cos(x_vals)
f2_prime_vals = 1 / x_vals


fig, axs = plt.subplots(2, 1, figsize=(10, 10))

axs[0].plot(x_vals, f1_vals, label=f"$ f(x) = {sp.latex(f1)} $", color='b')
axs[0].plot(x_vals, f1_prime_vals, label=f"$ f'(x) = {sp.latex(f1_prime)} $", color='g', linestyle="--")
axs[0].legend()

axs[1].plot(x_vals, f2_vals, label=f"$ f(x) = {sp.latex(f2)} $", color='b')
axs[1].plot(x_vals, f2_prime_vals, label=f"$ f'(x) = {sp.latex(f2_prime)} $", color='g', linestyle="--")
axs[1].legend()

plt.show()
```


3. Загальні витрати компанії від обсягу виробництва задані функцією $ f(x) = x^2 - 5x + 1 $, де $ x $ — кількість вироблених одиниць. Обчисліть похідну цієї функції з допомогою SymPy та візуалізуйте значення функції і її похідної на відрізку від 0 до 100.

```python
x = sp.symbols('x')
x_vals = np.linspace(0, 100)

f = x ** 2 - 5 * x + 1
f_prime = sp.diff(f)

print(f_prime)

f_vals = x_vals ** 2 - 5 * x_vals + 1
f_prime_vals = 2 * x_vals - 5

plt.figure(figsize=(10, 10))

plt.plot(x_vals, f_vals, label=f"$ f(x) = {sp.latex(f)} $", color="b")
plt.plot(x_vals, f_prime_vals, label=f"$ f'(x) = {sp.latex(f_prime)} $", color="g", linestyle="--")

plt.legend()
plt.show()
```
